#### 1. Importing Libraries

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as mpl
from sklearn.linear_model import LogisticRegression as LR
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder 
from sklearn.metrics import accuracy_score, precision_score
```

In the 'Class' column, value 0 represents legit transactions and value 1 represents fraudulent transactions. 

```python
data['Class'].value_counts()
```

We can observe that there are 492 fraudulent transactions. Therefore, it is an unbalanced dataset. 

We create separate data frames for normal and fraudulent transactions

```python
normal_transactions = data[data.Class==0]
fraud_transactions  = data[data.Class==1]
```

```python
print(normal_transactions.shape)
print(fraud_transactions.shape)
```

```python
normal_transactions['Amount'].describe()
```

```python
fraud_transactions['Amount'].describe()
```

```python
#Comparing values for both transactions
data.groupby('Class').mean()
```

This is a case of under-sampling. 
To solve this, we build a sample dataset containing similar distribution of both the classes. 

Number of fraudulent transactions: 492

Number of legit transactions:284315

We select 492 random transactions from the normal transactions dataset. 

```python
normal_sample = normal_transactions.sample(n=492)
```

```python
#Concatenating normal_sample and fraud_transactions to create one dataset. 
dataset=pd.concat([normal_sample,fraud_transactions],axis=0)
dataset
```

```python
dataset['Class'].value_counts()
```

```python
dataset.groupby('Class').mean()
```

### 4. Model Building and Training

##### Feature Extraction

```python
# X contains features i.e. all columns other than Class
# Y contains target variable i.e. Class

X=dataset.drop(columns='Class',axis=1)
Y=dataset['Class']
```

```python
X
```

```python
Y
```

##### Splitting the dataset into training and testing models

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=2,stratify=Y)
```

```python
print(f"X: {X.shape}, X_train:{X_train.shape},X_test:{X_test.shape}")
print(f"Y: {Y.shape}, Y_train:{Y_train.shape},Y_test:{Y_test.shape}")
```

```python
X_train.head()
```

```python
X_test.head()
```

#### Training the model on Logistic Regression

```python
model = LR()
```

```python
model.fit(X_train,Y_train)
```

### 5. Model Evaluation 

Using Accuracy score to determine the accuracy of the training model 

```python
training_prediction=model.predict(X_train)
print(training_prediction)
```

```python
training_accuracy = accuracy_score(Y_train,training_prediction)
print(f"The training accuracy is {training_accuracy}")
```

```python
testing_prediction = model.predict(X_test)
print(testing_prediction)
```

```python
testing_accuracy = accuracy_score(Y_test,testing_prediction)
print(f"The testing accuracy is {testing_accuracy}")
```

Our model shows accuracy of 90.863% on testing data
